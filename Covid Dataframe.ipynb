```python
# Lets find out missing values in specific columns : NewCases,NewDeaths,NewRecovered
df[df['NewCases'].isnull()]['NewCases']
```

```python
# Lets drop these columns with maximum NaN values
columns_to_drop = [ 'NewCases','NewDeaths','NewRecovered']

df.drop(columns=columns_to_drop,inplace = True)
```

```python
# Lets check if columns got dropped or not
df.head()
```

```python
# Lets check the count of null values now
missing_values_count = df.isnull().sum()

print(missing_values_count)
```

```python
# Lets drop Serious,Critical column
column_to_drop_one = ['Serious,Critical']
df.drop(columns = column_to_drop_one,inplace = True)
```

```python
# Lets print the column titles
df.columns
```

```python
# Again print the missing values and their sum
df.isnull().sum()
```

```python
# Lets replace missing values with mean in each remaining column
# Calculate mean of each column
# Convert DataFrame to numeric data type
df_numeric = df.apply(pd.to_numeric, errors='coerce')

# Calculate mean of each column
column_means = df_numeric.mean()

# Replace NaN values with column means
df.fillna(column_means, inplace=True)
```

```python
# Lets again print the missing values and their sum
df.isnull().sum()
```

```python
# Lets find the datatype of 'Continent' and 'WHO Region'
df.dtypes
```

```python
# Lets get the idea of those column content using df.head()
df.head()
```

```python
# Lets fill the missing values of those columns using mode
# Specify the column name for which you want to fill missing values with the mode
column_name = 'Continent'

# Calculate mode for the specified column
column_mode = df[column_name].mode()[0]

# Fill missing values in the specified column with the mode
df[column_name].fillna(column_mode, inplace=True)
```

```python
# Lets do same for the column name WHO Region
column_name = 'WHO Region'

# Calculate mode for the specified column
column_mode = df[column_name].mode()[0]

# Fill missing values in the specified column with the mode
df[column_name].fillna(column_mode, inplace=True)
```

```python
# Now lets check the missing values in dataframe
df.isnull().sum()
```

```python
# Finally there are zero null values in dataframe : Now lets relate the variable with scatterplot

# Finally, let's relate the variables with a scatterplot
sns.scatterplot(x="TotalCases", y="TotalRecovered", data=df)
plt.show()
```

```python
# Lets draw the same for total cases(x) vs total deaths(y)
sns.scatterplot(x="TotalCases", y="TotalDeaths", data=df)
plt.show()
```
